## Plotting with Pandas 
[Creating Plot with Pandas](https://pandas.pydata.org/docs/getting_started/intro_tutorials/04_plotting.html) <br>
🔑 Key Concepts to Remember
* **Built-in Plotting**: Pandas integrates with Matplotlib to provide easy .plot() functionality directly on Series and DataFrames.
* **Data Types Supported**: Works well with time series, categorical, and numerical data.
* **Kinds of Plots**: Common types include line, bar, histogram, area, scatter, box, and pie.
* **Chaining with Data Processing**: You can chain .plot() directly after filtering or aggregating data.<br>
* **Some common examples** 
1. df.plot(kind='line')         # Line plot (default)
2. df.plot(kind='bar')          # Bar chart
3. df.plot(kind='barh')         # Horizontal bar chart
4. df.plot(kind='hist')         # Histogram

### Scatter Plot 
A scatter plot displays values for two numerical variables as points on a 2D graph, helping to visualize the relationship or correlation between them (e.g., positive, negative, or none).
* X-axis: Independent variable
* Y-axis: Dependent variable
* Use Case: Ideal for spotting trends, clusters, and outliers in data.

```python
# example of scatter plot 
df.plot(kind='scatter',x='age',y='fare',color='red') # we can change color using hex code or color name 
```

### Example of subplots

```python
# example of subplot 
# Set the figure and axes (2 rows x 2 columns)
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Plot 1: Count of survivors
sns.countplot(data=df, x='survived', ax=axes[0, 0])
axes[0, 0].set_title('Survival Count')
axes[0, 0].set_xticklabels(['No', 'Yes'])

# Plot 2: Survival by gender
sns.countplot(data=df, x='sex', hue='survived', ax=axes[0, 1])
axes[0, 1].set_title('Survival by Gender')

# Plot 3: Age distribution
sns.histplot(data=df, x='age', bins=30, kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Age Distribution')

# Plot 4: Class distribution
sns.countplot(data=df, x='class', ax=axes[1, 1])
axes[1, 1].set_title('Passenger Class Distribution')

# Adjust layout
plt.tight_layout()
plt.show()
```

#### ✅ Key Features of Pandas Plotting (df.plot()):
* Quick & Easy Just one line of code needed to create basic plots like line, bar, histogram, area, pie, etc.
* Data-Aware : Automatically uses column names for axes and legend, making it intuitive with DataFrames.
* Multiple Plot Types Supports various chart types via kind parameter — 'line', 'bar', 'barh', 'hist', 'box', 'area', 'pie', 'scatter', etc.
* Subplots Support You can easily plot all columns as subplots using subplots=True.

## Plotting with Matplotlib
🔑 [Plotting With Matplotlib](https://matplotlib.org/stable/tutorials/pyplot.html) 

* Core Object: pyplot module (import matplotlib.pyplot as plt) is commonly used for basic plots.
* Figure & Axes: Everything is drawn on a Figure that contains one or more Axes (subplots).
* Customization: Almost every element (color, label, ticks, grid, legend, etc.) can be styled.

Rendering: Plots are rendered with plt.show() in most environments.

```python
## loading dataset
data=sns.load_dataset('tips')
```

### Pie Chart  
A pie chart is a circular chart divided into slices to illustrate proportions or percentages of a whole. Each slice represents a category’s contribution relative to the total.

* use the pie() function to draw pie charts:
* Useful for showing part-to-whole relationships.
* Works best with limited categories (usually under 6).
* Slices are often labeled with percentages.

```python
# creating pie charts using matplotlib
labels=data['sex'].unique()
plt.pie(data['sex'].value_counts())
plt.title('pie chart')
plt.legend(data['sex'].unique())    
```

Key Features of Plotting with matplotlib
* **Highly Customizable** - You can control every element of the plot—axes, colors, fonts, line styles, ticks, labels, legends, etc.
* **Supports Multiple Plot Types** - Line, bar, scatter, histogram, pie, 3D plots, and more. Ideal for both simple and complex visualizations.
* **Subplots and Layouts** - Easily create multiple plots in a single figure using plt.subplot() or plt.subplots() with grid layouts.
* **Interactive Plots (with %matplotlib notebook)** - Offers interactive plotting in Jupyter Notebooks and supports zooming and panning.
* Export in Multiple Formats - Save your plots as .png, .jpg, .svg, .pdf, etc., with high resolution using plt.savefig().

## Plotting With Seaborn 
🔑 [Plotting with Seaborn](https://seaborn.pydata.org/tutorial/introduction.html)
Seaborn is a visualization library built on top of matplotlib. It provides a high-level interface for
creating attractive and informative statistical graphics.
* **Built-in Themes**: Seaborn comes with beautiful default themes for better aesthetics (e.g., darkgrid, whitegrid, etc.).
* **Statistical Plots**: Seaborn focuses on visualizing statistical relationships, distributions, and categorical comparisons.
* **Automatic Estimation**: Many Seaborn plots, like regression plots, automatically fit statistical models (e.g., linear regression) to your data.


### Box Plot
A box plot is a graphical representation of the distribution of a dataset. It is a useful tool for
visualizing the five-number summary of a dataset: the minimum value, the first quartile (Q 1), the median (Q 2), the third quartile (Q 3), and the maximum value.

```python
sns.boxplot(x='sex',y='age',data=df,hue='class')
```
